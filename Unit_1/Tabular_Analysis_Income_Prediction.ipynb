```python
from fastai.tabular.all import *

# Télécharge et décompresse les données
path = untar_data(URLs.ADULT_SAMPLE)

# Charge les données tabulaires
dls = TabularDataLoaders.from_csv(
    path/'adult.csv',
    path=path,
    y_names='salary',
    cat_names=['workclass', 'education', 'marital-status', 'occupation', 'relationship'],
    cont_names=['age', 'fnlwgt', 'education-num'],
    procs=[Categorify, FillMissing, Normalize]
)
```

```python
#see the data
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
# Train the model using the 1-cycle policy for 2 epochs
learn.fit_one_cycle(2)
# - 2: Number of epochs (complete passes through the training dataset)
```
